```python
actual_acc = accuracy
scores_df = pd.DataFrame(columns = ['removed_feature','accuracy_score', 'difference'])
index = 0
for feature in all_features:
    partial_num = numerical.copy()
    partial_cat = categorical.copy()
    if (feature in numerical):
        partial_num.remove(feature)
    else :
        partial_cat.remove(feature)
    
    # dv = DictVectorizer(sparse=False)
    # train_dict = df_train[partial_features].to_dict(orient='records')
    # X_train = dv.fit_transform(train_dict)

    scaler = StandardScaler()
    X_train_numerical = scaler.fit_transform(df_train[partial_num].values)

    ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
    X_train_categorical = ohe.fit_transform(df_train[partial_cat].values)
    X_train = np.column_stack([X_train_categorical, X_train_numerical])

    model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)

    # dict_val = df_val[partial_features].to_dict(orient='records')
    # X_val = dv.transform(dict_val)

    X_val_numerical = scaler.fit_transform(df_val[partial_num].values)
    X_val_categorical = ohe.fit_transform(df_val[partial_cat].values)
    X_val = np.column_stack([X_val_numerical, X_val_categorical])

    y_pred = model.predict(X_val)
    acc_score = accuracy_score(y_pred, y_val)

    scores_df.loc[index] = [feature, acc_score, actual_acc - acc_score]
    index += 1

# for feature in categorical:
#     partial_features = numerical.copy()
#     partial_features.remove(feature)

#     ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
#     X_train_cate
```

```python

for reg_value in [0.01, 0.1, 1, 10, 100]:
    model = Ridge(alpha=reg_value, solver='sag', max_iter=1000, random_state=42)
    model.fit(X_train,y_train)

    y_pred = model.predict(X_val)
    # print(y_pred)
    accuracy = (y_val == (y_pred >= 0.5)).mean()
    acc_score = np.round(accuracy,3)

    print(f' Alpha : {reg_value}, Accuracy Score : {acc_score}')
    
    
```
